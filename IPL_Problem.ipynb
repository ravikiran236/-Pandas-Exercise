```python
# creating and adding strike rate column to the data frame
ipl['strike rate'] = (ipl['runs']/ ipl['balls'])*100
```

```python
#3 list down the century makers, their scores and their strike rates
ipl[['batsman_name', 'runs', 'strike rate']][ipl['runs']>= 100]
```

```python
#6 what are the total number of duck dismissals?
ipl[ipl['runs'] ==0].count()['runs']
```

```python
#list down the top 5 most consistent batsmen
#find out the cases where a batsman is not out or did not bat
ipl['out_status'] = ipl.dismissal_mode.map(lambda x:0 if (x=='not out'or x=='did not bat') else 1)
#ipl.head(20)
#ipl[ipl.batsman_name.str.contains('Rohit')]
#create group on batsman name
grp_player = ipl.groupby('batsman_name')
#aggregate the runs and out_status columns
ipl1 = grp_player['runs','out_status'].agg(np.sum)
ipl1
#calculate average
ipl1['average'] = ipl1.runs/ipl1.out_status
#print top 5 batsmen (exclude the cases where total count of out_cases is 0)
ipl1[ipl1.out_status>=5].sort_values(by='average',ascending=0).head(5)
```

```python
# Creating Boundary Percentage field
ipl['boundary %'] = (((4*ipl['fours'])+(6*ipl['sixes']))/ipl['runs'])*100
```

```python
# Checking whether Boundary % field is created or not
ipl.head()
```

```python
# lit top 5 players by contribution of boundaries to total scores
ipl.nlargest(5,{'boundary %'},keep = 'first')
```

```python
# which location is most batsman friendly
# the location which has the most runs made per match + where %of runs from boundaries is high
grp_location = ipl.groupby('match_location')
#grp_location.groups
ipl['boundary_runs'] = ipl['fours']*4 + ipl['sixes']*6
ipl1 = grp_location['runs','boundary_runs'].agg(np.sum)
ipl2 = grp_location.match_description.nunique()
ipl1['match_count'] = ipl2
ipl1['runs_per_match'] = ipl1.runs/ipl1.match_count
ipl1['boundary_contrib'] = (ipl1.boundary_runs/ipl1.runs)*100
ipl1
```
